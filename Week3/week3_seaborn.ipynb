```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


data = sns.load_dataset("flights") 
data = data.pivot("month", "year", "passengers")

# Heatmap   annot -> Visibility of numbers in each cell
#           fmt ->  d means make integer the numbers in each cell
#           cmap -> Defines the color
sns.heatmap(data=data, annot=True, fmt="d", cmap="YlGnBu")

plt.show()
```

```python
# Line plot
sns.lineplot(data=data, x="year", y="passengers")

plt.show()
```

```python
#Another data
data = sns.load_dataset('iris')

#Swarm plot
sns.swarmplot(x="species", y="petal_length", data=data) # spiiz

plt.show()
```

```python
#Another data
data = sns.load_dataset("titanic")

#Bar plot
sns.countplot(data=data, x="class")

plt.show()
```

```python
#Another data
data = sns.load_dataset("tips")

# Scatter plot
sns.scatterplot(data=data, x="total_bill", y="tip")
plt.show()
```

```python
#Another data
data = sns.load_dataset("titanic")

# Facet grid
g = sns.FacetGrid(data, col="sex")

g.map(sns.histplot, "age")
plt.show()
```

```python
#Iris dataset
iris = sns.load_dataset("iris")

#Data is grouped according to the Species column
g = sns.clustermap(iris.drop("species", axis=1), cmap="coolwarm", standard_scale=1)

plt.show()
```
