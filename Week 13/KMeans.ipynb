```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('./column_2C_weka.csv')
plt.scatter(df['pelvic_radius'],df['degree_spondylolisthesis'])
plt.xlabel('pelvic_radius')
plt.ylabel('degree_spondylolisthesis')
plt.show()
```

```python
# KMeans Clustering
df2 = df.loc[:,['degree_spondylolisthesis','pelvic_radius']]

#This number means compare two closest data with our data
kmeans = KMeans(n_clusters = 2)

kmeans.fit(df2)
labels = kmeans.predict(df2)
plt.scatter(df['pelvic_radius'],df['degree_spondylolisthesis'],c = labels)
plt.xlabel('pelvic_radius')
plt.xlabel('degree_spondylolisthesis')
plt.show()
```
