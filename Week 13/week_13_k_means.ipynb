```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# It generates 1000 values with 5 sigma at 25% average. 66@% of the data is between (25+5)30 and (25-5)20. This is called Gaussian (normal) distribution
x1 = np.random.normal(25,5,1000)
y1 = np.random.normal(25,5,1000)

x2 = np.random.normal(55,5,1000)
y2 = np.random.normal(60,5,1000)

x3 = np.random.normal(55,5,1000)
y3 = np.random.normal(15,5,1000)
```

```python
# X features merge on the X axis and y features merge on the y axis
x= np.concatenate((x1,x2,x3), axis=0)

y= np.concatenate((y1,y2,y3), axis=0)
```

```python
dictionary = {"x":x, "y":y}
```

```python
data = pd.DataFrame(dictionary)
```

```python
plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.scatter(x3,y3)
plt.show()
```

```python
plt.scatter(x1,y1, color = 'black')
```

```python
wcss = []

for i in range(1,15):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
 
plt.plot(range(1,15), wcss)
plt.xlabel("number of k (cluster) value") 
plt.ylabel("wcss")
plt.title("Elbow Method")
plt.grid()
plt.show()
```

```python
# For k= 2

kmeans2 = KMeans(n_clusters=2)

clusters = kmeans2.fit_predict(data)

data['label']= clusters

plt.scatter(data.x[data.label==0],data.y[data.label == 0],color = 'red')
plt.scatter(data.x[data.label==1],data.y[data.label == 1],color = 'green')
plt.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],color = 'blue')
```

```python
# For k= 3

kmeans2 = KMeans(n_clusters=3)

clusters = kmeans2.fit_predict(data)

data['label']= clusters

plt.scatter(data.x[data.label==0],data.y[data.label == 0],color = 'red')
plt.scatter(data.x[data.label==1],data.y[data.label == 1],color = 'green')
plt.scatter(data.x[data.label==2],data.y[data.label == 2],color = 'blue')
plt.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],color = 'yellow')
```
